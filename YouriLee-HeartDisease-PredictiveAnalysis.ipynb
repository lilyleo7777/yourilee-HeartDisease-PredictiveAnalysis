```python
def ctg_dist(df, cols, hue=None, rows=2, columns=3):
    fig, axes = plt.subplots(rows, columns, figsize=(12, 7))
    axes = axes.flatten()
    for i, j in zip(cols, axes):
        if hue is None:
            sns.countplot(x=i, data=df, palette='Paired', hue=i, ax=j, order=df[i].value_counts().index, legend=False)
        else:
            sns.countplot(x=i, data=df, palette='Paired', hue=hue, ax=j, order=df[i].value_counts().index)
        j.tick_params(labelrotation=0)
        total = float(len(df[i]))
        j.set_title(f'{str(i)} Distribution')
        for p in j.patches:
            height = p.get_height()
            j.text(p.get_x() + p.get_width() / 2., height + 2, '{:.2f}%'.format((height / total) * 100), ha='center')
    plt.tight_layout()
```

```python
from scipy import stats
# Split the data into two groups based on 'HeartDisease'
group1 = chol_not0[chol_not0['HeartDisease'] == 0]['Cholesterol']
group2 = chol_not0[chol_not0['HeartDisease'] == 1]['Cholesterol']
# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
# Check if the p-value is less than 0.05 (common significance level)
if p_value < 0.05:
    print("The difference between the two groups is statistically significant.")
else:
    print("The difference between the two groups is not statistically significant.")
```

#### **Categorical Features**

Changing cateogrical values into numerical values

RestingECG - Nominal
  one hot encoded

Gender - Nominal
  M = 0, F = 1

Exercise Angia - Nominal
  N = 0, Y = 1

ST_Slope - ordinal
  Up = 0, Flat = 1, Down = 2

ChestPainType - Ordinal
  ASY = 0, NAP = 1, ATA = 2, TA = 3

Why ST_Slope is Ordinal:
There is a natural order to the ST_Slope categories in terms of the severity or risk they represent:
Upsloping: Often considered less indicative of ischemia.
Flat: More concerning than upsloping, indicating potential ischemia.
Downsloping: Most concerning, strongly indicative of myocardial ischemia.

Why RestingECG is Nominal:
There is no inherent order or ranking among the categories. They represent distinct types of ECG findings, but one is not inherently "greater" or "less" than the other in a way that creates a natural sequence.
The categories are simply different conditions observed in the resting ECG.

Why ChestPainType is Ordinal:
There is an inherent order or ranking among the categories in terms of severity or typicality of angina:
Typical Angina: Most indicative of heart-related issues.
Atypical Angina: Less indicative of heart-related issues.
Non-anginal Pain: Least indicative of heart-related issues.
Asymptomatic: No chest pain, indicating the absence of symptoms.

```python
chestPain_mapping = {'ASY': 0, 'NAP': 1, 'ATA': 2, 'TA': 3}
df_p.loc[:,'ChestPainType'] = df_p['ChestPainType'].map(chestPain_mapping)

gender_map = {'M': 0, 'F': 1}
df_p.loc[:,'Gender'] = df_p['Gender'].map(gender_map)

ex_ang_map = {'N': 0, 'Y': 1}
df_p.loc[:,'ExerciseAngina'] = df_p['ExerciseAngina'].map(ex_ang_map)

ST_slope_map = {'Up': 0, 'Flat': 1, 'Down':2}
df_p.loc[:,'ST_Slope'] = df_p['ST_Slope'].map(ST_slope_map)

df_p_test.loc[:,'ChestPainType'] = df_p_test['ChestPainType'].map(chestPain_mapping)
df_p_test.loc[:,'Gender'] = df_p_test['Gender'].map(gender_map)
df_p_test.loc[:,'ExerciseAngina'] = df_p_test['ExerciseAngina'].map(ex_ang_map)
df_p_test.loc[:,'ST_Slope'] = df_p_test['ST_Slope'].map(ST_slope_map)
```
